```python
import pandas as pd
import os
import glob

# Define the source and destination directories
source_dir = 'D:\\Parkinson\'s\\Filtered Data'
destination_dir = 'D:\\Parkinson\'s\\Filtered_Data_9_to_5'

# Create the destination directory if it doesn't exist
if not os.path.exists(destination_dir):
    os.makedirs(destination_dir)

# Get a list of all CSV files in the source directory
csv_files = glob.glob(os.path.join(source_dir, '*.csv'))

# Process each CSV file
for file_path in csv_files:
    # Read the CSV file into a DataFrame
    df = pd.read_csv(file_path)

    # Convert the 'Datetime_PST' column to datetime, inferring the datetime format
    df['Datetime_PST'] = pd.to_datetime(df['Datetime_PST'], errors='coerce', infer_datetime_format=True)

    # Drop rows where 'Datetime_PST' could not be converted
    df = df.dropna(subset=['Datetime_PST'])

    # Filter the DataFrame to include only times between 9 AM and 5 PM PST
    df_filtered = df[df['Datetime_PST'].dt.time.between(pd.Timestamp('09:00:00').time(),
                                                         pd.Timestamp('17:00:01').time())]

    # Construct the path for the filtered file to be saved
    base_name = os.path.basename(file_path)
    filtered_file_path = os.path.join(destination_dir, base_name)

    # Save the filtered DataFrame to a new CSV file in the destination directory
    df_filtered.to_csv(filtered_file_path, index=False)

print("All files have been processed and saved to the 'Filtered_Data_9_to_5' directory.")
```

# THE Filtered_Data_9_to_5 was updated as there was issues with the Datetime_PST for some of the files which was leading to lower number of features while doing TSFEL

```python
import pandas as pd
import os

# Define the source and destination directories
source_dir = 'D:\\Parkinson\'s\\Filtered Data'
destination_dir = 'D:\\Parkinson\'s\\Filtered_Data_9_to_5_updated'

# Create the destination directory if it doesn't exist
os.makedirs(destination_dir, exist_ok=True)

# List all the .csv files in the source directory
csv_files = [file for file in os.listdir(source_dir) if file.endswith('.csv')]

# Process each file
for file_name in csv_files:
    # Define file paths
    file_path = os.path.join(source_dir, file_name)
    destination_file_path = os.path.join(destination_dir, file_name)

    # Read the CSV file into a DataFrame
    df = pd.read_csv(file_path)

    # Convert the 'Datetime_PST' column to datetime, inferring the datetime format
    df['Datetime_PST'] = pd.to_datetime(df['Datetime_PST'], format='%Y-%m-%d %H:%M:%S.%f%z', errors='coerce')

    # Drop rows where 'Datetime_PST' could not be converted
    df = df.dropna(subset=['Datetime_PST'])

    # Filter the DataFrame to include only times between 9 AM and 5 PM PST
    df_filtered = df[df['Datetime_PST'].dt.time.between(pd.Timestamp('09:00:00').time(),
                                                        pd.Timestamp('17:00:01').time())]

    # Format the 'Datetime_PST' column to include milliseconds before saving
    df_filtered['Datetime_PST'] = df_filtered['Datetime_PST'].dt.strftime('%Y-%m-%d %H:%M:%S.%f%z')

    # Save the filtered DataFrame to a new CSV file in the specified destination
    df_filtered.to_csv(destination_file_path, index=False)

# Load the first processed file to show the first 5 rows
first_processed_file_path = os.path.join(destination_dir, csv_files[0]) if csv_files else ""
if first_processed_file_path:
    first_df = pd.read_csv(first_processed_file_path)
    display = first_df.head(5)
else:
    display = "No files processed."

display
```
